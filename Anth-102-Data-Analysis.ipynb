```python
import csv
import numpy as np
import scipy.optimize
import random
import json
import urllib.request
import sklearn
import string
from collections import defaultdict
import urllib.parse, urllib.error
import nltk
from nltk.stem.porter import *
from sklearn import linear_model
import math
from scipy import spatial
from sklearn.metrics import mean_squared_error
from sklearn import svm
import matplotlib.pyplot as plt
from collections import Counter

with open('Anth102-communities-and-entertainment-survey.csv') as csvfile:
    readCSV = csv.reader(csvfile, delimiter=',')
    countMusicAlone = 0
    countMusicFriends = 0
    
    countFriendAlone = 0
    countFriendFriends = 0
    for row in readCSV:
        #print(row[8])
        if(row[1] == "You like the music being presented" and row[8] == "Yes"):
            countMusicAlone += 1
        if(row[1] == "You like the music being presented" and row[8] == "No"):
            countMusicFriends += 1
            
            
        if(row[1] == "Your friends are going" and row[8] == "Yes"):
            countFriendAlone += 1
        if(row[1] == "Your friends are going" and row[8] == "No"):
            countFriendFriends += 1
        
        
    print("count music alone: " + str(countMusicAlone))
    print("count music friends: " + str(countMusicFriends))
    print("")
    print("count friend alone: " + str(countFriendAlone))
    print("count friend friends: " + str(countFriendFriends))
        
    X = ['Music Alone', 'Music Friends', 'Friend Alone', 'Friend Friends']
    Y = [countMusicAlone, countMusicFriends, countFriendAlone, countFriendFriends]

    plt.bar(X, Y)
    plt.show()
```
